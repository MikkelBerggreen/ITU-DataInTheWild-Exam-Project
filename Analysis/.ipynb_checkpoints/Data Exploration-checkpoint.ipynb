```python
import pandas as pd
import numpy as np
import json
import seaborn as sns
import matplotlib.pyplot as plt
```

## Parsing and cleaning

```python
with open('data.json', encoding="utf-8") as json_file:
    data = json.load(json_file)
df = pd.json_normalize(data)
```

```python
df.dtypes
```

```python
nan_count = df.isna().sum()

print(nan_count)
```

```python
df
```

```python
# drop nan values because ....
df = df.dropna()
```

```python
df["CapitalCurrency"] = [x.split()[1] if x != None else "DKK" for x in df["RegisteredCapital"]]
df["RegisteredCapital"] = [float(x.split()[0].replace(".", "").replace(",", ".")) if x != None else 0 for x in df["RegisteredCapital"]]
```

```python
labels = ["Other"]
counts = [0]

for industry in df["IndustryName"].unique():
    if len(df[df['IndustryName'] == industry]) < 20:
        counts[0] = counts[0] +len(df[df['IndustryName'] == industry])
    else:
        labels.append(industry)
        counts.append(len(df[df['IndustryName'] == industry]))
plt.rcParams["figure.figsize"] = (16,12)
plt.pie(counts, labels=labels, autopct='%1.0f%%', pctdistance=1.1, labeldistance=1.2)
```
